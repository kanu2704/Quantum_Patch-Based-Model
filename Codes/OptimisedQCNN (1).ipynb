```python
# Required installations:
# pip install qiskit==1.4.2 tensorflow numpy pandas matplotlib pennylane silence-tensorflow
# pip install pennylane-qiskit qiskit-ibm-runtime qiskit-aer

import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import warnings
import cv2
import numpy as np
tf.keras.backend.set_floatx('float32')
warnings.filterwarnings('ignore', category=UserWarning, module='tensorflow')
from silence_tensorflow import silence_tensorflow
silence_tensorflow()

# Quantum imports
import pennylane as qml
from pennylane import numpy as pnp
import math

# Quantum circuit setup - Use fast local simulator for development
n_qubits = 4
patch_size = 2
layers = 1
levels = 2

# Use the fastest local simulator
print("Using optimized local quantum simulator for faster development...")
dev = qml.device('default.qubit', wires=n_qubits)

# Optimized quantum circuit with batch processing capability
@qml.qnode(dev, interface="tf")
@tf.autograph.experimental.do_not_convert
def quantum_circuit(inputs, weights):
    """Optimized quantum circuit for fast classical execution"""
    # Feature mapping
    for i in range(n_qubits):
        angle = tf.clip_by_value(pnp.pi * tf.cast(inputs[i], tf.float64), -2*pnp.pi, 2*pnp.pi)
        qml.RY(angle, wires=i)

    # Variational layers
    for l in range(layers):
        for i in range(n_qubits):
            qml.RY(tf.cast(weights[l][i], tf.float64), wires=i)
        for i in range(n_qubits):
            qml.CNOT(wires=[i, (i + 1) % n_qubits])

    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# Load and preprocess MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

def resize_images(images, size=(12, 12)):
    """Vectorized image resizing for speed"""
    resized_images = np.array([cv2.resize(img, size, interpolation=cv2.INTER_AREA) for img in images])
    return resized_images

print("Resizing images...")
train_images = resize_images(train_images)
test_images = resize_images(test_images)

# Function to select N samples for each of given classes
def get_multiple_classes(images, labels, classes, num_samples):
    """Fetch `num_samples` for each class in `classes`"""
    all_images = []
    all_labels = []
    for cls in classes:
        indices = np.where(labels == cls)[0][:num_samples]
        all_images.append(images[indices])
        all_labels.append(labels[indices])
    return np.concatenate(all_images), np.concatenate(all_labels)

# Selecting 250 samples per digit (4–5) for training and testing
print("Selecting training samples for digits 0 to 5...")
train_images_selected, train_labels_selected = get_multiple_classes(
    train_images, train_labels, classes=[4, 5], num_samples=500
)

print("Selecting test samples for digits 0 to 5...")
test_images_selected, test_labels_selected = get_multiple_classes(
    test_images, test_labels, classes=[4, 5], num_samples=100
)

# Shuffle training data
shuffle_indices = np.random.permutation(len(train_images_selected))
train_images_selected = train_images_selected[shuffle_indices]
train_labels_selected = train_labels_selected[shuffle_indices]

# Normalize
train_images_selected = train_images_selected / 255.0
train_images_selected = tf.cast(train_images_selected, tf.float32)

test_images_selected = test_images_selected / 255.0
test_images_selected = tf.cast(test_images_selected, tf.float32)

print(f"Training data shape: {train_images_selected.shape}")
print(f"Test data shape: {test_images_selected.shape}")

# MAJOR OPTIMIZATION: Pre-extract all patches for all images
def extract_all_patches_vectorized(images, patch_size):
    """Vectorized patch extraction for all images at once"""
    batch_size, height, width = images.shape

    all_patches = []
    for img in images:
        img_patches = []
        for i in range(0, height, patch_size):
            for j in range(0, width, patch_size):
                if i + patch_size <= height and j + patch_size <= width:
                    patch = img[i:i+patch_size, j:j+patch_size]
                    img_patches.append(tf.reshape(patch, [-1]))
        all_patches.append(tf.stack(img_patches))

    return tf.stack(all_patches)

# Pre-extract patches for ALL images (training and test)
print("Pre-extracting patches for all training images...")
train_patches_level1 = extract_all_patches_vectorized(train_images_selected, patch_size)
print(f"Level 1 training patches shape: {train_patches_level1.shape}")

print("Pre-extracting patches for all test images...")
test_patches_level1 = extract_all_patches_vectorized(test_images_selected, patch_size)
print(f"Level 1 test patches shape: {test_patches_level1.shape}")

# Initialize quantum weights
tf.random.set_seed(42)
pnp.random.seed(42)
weight_scale = 0.1

weights_l1 = tf.Variable(tf.random.normal((layers, n_qubits), stddev=weight_scale), trainable=True)
weights_l2 = tf.Variable(tf.random.normal((layers, n_qubits), stddev=weight_scale), trainable=True)

# Classical model - input will be 9 (3x3 flattened)
classical_model = tf.keras.Sequential([
    tf.keras.Input(shape=(9,)),
    tf.keras.layers.Dense(10, activation='softmax')
])

optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()

# Process quantum circuits efficiently
def process_patches_batch(patches, weights, max_patches=36):
    """Process multiple patches efficiently"""
    if patches.shape[0] > max_patches:
        patches = patches[:max_patches]

    outputs = []
    for i in range(patches.shape[0]):
        patch = patches[i]
        if patch.shape[0] < n_qubits:
            patch = tf.concat([patch, tf.zeros(n_qubits - patch.shape[0])], axis=0)
        else:
            patch = patch[:n_qubits]

        q_output = quantum_circuit(patch, weights)
        if isinstance(q_output, (list, tuple)):
            q_output = tf.stack(q_output)
        q_output = tf.cast(tf.math.real(q_output), tf.float32)
        outputs.append(tf.reduce_mean(q_output))

    return tf.stack(outputs)

def forward_two_levels(img_idx, patches_level1):
    """Two-level forward pass with pre-extracted patches"""
    try:
        patches1 = patches_level1[img_idx]
        q_outputs1 = process_patches_batch(patches1, weights_l1, max_patches=36)

        if len(q_outputs1) > 36:
            q_outputs1 = q_outputs1[:36]
        elif len(q_outputs1) < 36:
            padding = 36 - len(q_outputs1)
            q_outputs1 = tf.concat([q_outputs1, tf.zeros(padding)], axis=0)

        q_outputs1_reshaped = tf.reshape(q_outputs1, (6, 6))

        # Level 2: Extract 2x2 patches from 6x6 grid to get 3x3 output
        patches2 = []
        for i in range(0, 6, patch_size):
            for j in range(0, 6, patch_size):
                if i + patch_size <= 6 and j + patch_size <= 6:
                    patch = q_outputs1_reshaped[i:i+patch_size, j:j+patch_size]
                    patches2.append(tf.reshape(patch, [-1]))

        patches2 = tf.stack(patches2)
        q_outputs2 = process_patches_batch(patches2, weights_l2, max_patches=9)

        if len(q_outputs2) > 9:
            q_outputs2 = q_outputs2[:9]
        elif len(q_outputs2) < 9:
            padding = 9 - len(q_outputs2)
            q_outputs2 = tf.concat([q_outputs2, tf.zeros(padding)], axis=0)

        q_outputs2_flattened = tf.reshape(q_outputs2, (1, -1))
        logits = classical_model(q_outputs2_flattened)
        return logits

    except Exception as e:
        print(f"Error in forward pass: {e}")
        return tf.zeros((1, 10))

# Optimized training step
@tf.function
def train_step_optimized(img_idx, y_label):
    with tf.GradientTape() as tape:
        predictions = forward_two_levels(img_idx, train_patches_level1)
        loss = loss_fn(tf.reshape(y_label, (1,)), predictions)

    trainable_vars = [weights_l1, weights_l2] + classical_model.trainable_variables
    gradients = tape.gradient(loss, trainable_vars)
    gradients = [tf.clip_by_norm(g, 1.0) if g is not None else g for g in gradients]
    optimizer.apply_gradients(zip(gradients, trainable_vars))
    return loss, predictions

# Training loop
print("Starting optimized training with 2 levels...")
print("=" * 50)

epochs = 10
batch_losses = []
start_time = tf.timestamp()

for epoch in range(epochs):
    epoch_loss = 0.0
    correct_predictions = 0
    shuffle_indices = np.random.permutation(len(train_images_selected))
    print(f"Epoch {epoch+1}/{epochs}")

    for i, idx in enumerate(shuffle_indices):
        try:
            label = train_labels_selected[idx]
            loss, predictions = train_step_optimized(idx, tf.constant(label))
            epoch_loss += loss.numpy()
            predicted_class = tf.argmax(predictions, axis=1).numpy()[0]
            if predicted_class == label:
                correct_predictions += 1
            if (i + 1) % 10 == 0:
                print(f"  Processed {i+1}/{len(shuffle_indices)} samples")
        except Exception as e:
            print(f"Error processing sample {i}: {e}")
            continue

    avg_loss = epoch_loss / len(shuffle_indices)
    accuracy = correct_predictions / len(shuffle_indices)
    batch_losses.append(avg_loss)
    print(f"Epoch {epoch+1} | Loss: {avg_loss:.6f} | Accuracy: {accuracy:.4f}")

end_time = tf.timestamp()
training_time = end_time - start_time
print(f"\nTraining completed in {training_time:.2f} seconds!")

# SUPER-FAST TEST METHOD WITH PRE-COMPUTED LEVEL 1 OUTPUTS
def test_model_super_fast():
    print("Running super-fast test with pre-computed level 1 outputs...")
    correct = 0
    total = len(test_images_selected)

    print("Pre-computing all level 1 quantum outputs...")
    precompute_start = tf.timestamp()
    all_level1_outputs = []
    for i in range(total):
        patches1 = test_patches_level1[i]
        q_outputs1 = process_patches_batch(patches1, weights_l1, max_patches=36)
        if len(q_outputs1) > 36:
            q_outputs1 = q_outputs1[:36]
        elif len(q_outputs1) < 36:
            padding = 36 - len(q_outputs1)
            q_outputs1 = tf.concat([q_outputs1, tf.zeros(padding)], axis=0)
        all_level1_outputs.append(q_outputs1)
        if (i + 1) % 20 == 0:
            print(f"  Pre-computed {i+1}/{total} level 1 outputs")
    precompute_end = tf.timestamp()
    precompute_time = precompute_end - precompute_start

    print("Processing level 2 and classification...")
    inference_start = tf.timestamp()
    for i in range(total):
        try:
            q_outputs1_reshaped = tf.reshape(all_level1_outputs[i], (6, 6))
            patches2 = []
            for row in range(0, 6, patch_size):
                for col in range(0, 6, patch_size):
                    if row + patch_size <= 6 and col + patch_size <= 6:
                        patch = q_outputs1_reshaped[row:row+patch_size, col:col+patch_size]
                        patches2.append(tf.reshape(patch, [-1]))
            patches2 = tf.stack(patches2)
            q_outputs2 = process_patches_batch(patches2, weights_l2, max_patches=9)
            if len(q_outputs2) > 9:
                q_outputs2 = q_outputs2[:9]
            elif len(q_outputs2) < 9:
                padding = 9 - len(q_outputs2)
                q_outputs2 = tf.concat([q_outputs2, tf.zeros(padding)], axis=0)
            q_outputs2_flattened = tf.reshape(q_outputs2, (1, -1))
            predictions = classical_model(q_outputs2_flattened)
            predicted_class = tf.argmax(predictions, axis=1).numpy()[0]
            if predicted_class == test_labels_selected[i]:
                correct += 1
        except Exception as e:
            print(f"Error in sample {i}: {e}")
            continue
    inference_end = tf.timestamp()
    inference_time = inference_end - inference_start
    total_test_time = precompute_time + inference_time
    accuracy = correct / total
    print(f"\nSuper-fast Test Results:")
    print(f"  Pre-compute time: {precompute_time:.2f} seconds")
    print(f"  Inference time: {inference_time:.2f} seconds")
    print(f"  Total test time: {total_test_time:.2f} seconds")
    print(f"  Test Accuracy: {accuracy:.4f} ({correct}/{total})")
    return accuracy, total_test_time, precompute_time, inference_time

print("\n" + "="*60)
print("RUNNING SUPER-FAST TEST")
print("="*60)
test_accuracy, total_test_time, precompute_time, inference_time = test_model_super_fast()

# Plot results
plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
plt.plot(batch_losses)
plt.title('Training Loss (2 Levels)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)

plt.subplot(1, 4, 2)
plt.bar(['Training', 'Test'], [batch_losses[-1], test_accuracy])
plt.title('Performance Metrics')
plt.ylabel('Value')

plt.subplot(1, 4, 3)
phases = ['Pre-compute\nLevel 1', 'Level 2 +\nClassification']
times = [precompute_time, inference_time]
plt.bar(phases, times)
plt.title('Test Phase Breakdown')
plt.ylabel('Time (seconds)')

plt.subplot(1, 4, 4)
plt.text(0.1, 0.9, f"Training Time: {training_time:.1f}s", transform=plt.gca().transAxes, fontsize=11, weight='bold')
plt.text(0.1, 0.8, f"Total Test Time: {total_test_time:.1f}s", transform=plt.gca().transAxes, fontsize=11, weight='bold')
plt.text(0.1, 0.7, f"Test Accuracy: {test_accuracy:.4f}", transform=plt.gca().transAxes, fontsize=11)
plt.axis('off')
plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("OPTIMIZATION SUMMARY")
print("="*60)
print("✓ Pre-extracted patches for all images")
print("✓ Pre-computed level 1 quantum outputs")
print("✓ Minimized quantum circuit calls")
print("✓ Two-level hierarchical processing: 12x12 → 6x6 → 3x3")
print(f"✓ Total training time: {training_time:.1f} seconds")
print(f"✓ Total test time: {total_test_time:.1f} seconds")
```
